## JointGrid:
----

```python
g3 = sns.JointGrid(data=df, x='alcohol', y='residual sugar')
g3.plot(sns.regplot, sns.distplot)
```

```python
g3 = sns.JointGrid(data=df, x='alcohol', y='residual sugar')
g3.plot(sns.regplot, sns.distplot)
g3.plot_joint(sns.kdeplot)
g3.plot_marginals(sns.kdeplot, shade=True)
```

```python
sns.jointplot(data=df, x='alcohol', y='residual sugar', kind='hex', order=)
```

```python
g = (sns.jointplot(x="alcohol",
             y="pH",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
```

```python
g = (sns.jointplot(x="alcohol",
             y="density",
             kind='scatter',
             data=df,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
    
plt.show()
```
