## Relational plots and subplots:


1. Examples of relational plots(scatter and line):
 - Height vs. weight
 - Number of school absences vs. final grade
 - GDP vs. percent literate

2. Scatter plot 
 - shows relationship between two quantitative variables.
 - Each plot point is an independent observation.

3. Line plot
 - Each plot point represents the same 'thing,' typically tracked over time. 
 - shows the shaded region which represents 95% of confident interval assuming the dataset is random sample.

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips,kind='line' )
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, col='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, row='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, row='smoker', col='time')
```

```python
sns.set_palette('RdBu')
g=sns.relplot(x='total_bill', y='tip', data=tips, col='smoker')
g.fig.suptitle('Tips by total bill')
g.set_titles('Smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', size='size')
```

## Point plots:

1. Points show mean of quantitative variable
2. Vertical lines show 95% cofidence intervals

### Point plots vs. line plots:

- Both show:
    1. mean of quantitative variable
    2. 95% confidence intervals for the mean

- Differences:
    1. Line plot has quantitative variable (usually time) on x-axis
    2. Point plot has categorical variable on x-axis

### Point plots vs. bar plots:

- Both show:
    1. mean of quantitative variable
    2. 95% confidence intervals for the mean

- Differences:
    1.

## FacetGrid vs. AxesSubplot

### FacetGrid: can create subplots
1. relplot()
2. catplot()
3. example: g.fig.suptitle()

### AxesSubplot: only create a single plot
1. scatterplot()
2. countplot()
3. example: g.set_title()

```python
g= sns.catplot(kind='bar', data=df_jan_2019_first_seven, x='nation', y='visitor')
g.fig.suptitle('Jan_2019_first_seven_countries')
plt.xticks(rotation=45)
```

```python
s= sns.catplot(x='nation', y='visitor', data=df_jan_2019, kind='bar')
s.fig.suptitle('Jan_2019_visitors_by countries', y=1.05)
plt.xticks(rotation=45 )
```
