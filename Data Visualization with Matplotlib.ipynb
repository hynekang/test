```python
def plot_timeseries(axes, x,y, color1, xlabel, ylabel, color2,color3):
    axes.plot(x,y, color=color1)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color2)
    axes.tick_params('y',colors=color3)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax,tem_south_korea_since_2008.index, tem_south_korea_since_2008['avg_temp_c'],'pink','Time since 2008','Temperture in C','#39cc19','#cf3797')

ax2=ax.twinx()
plot_timeseries(ax2,tem_south_korea_since_2008.index, tem_south_korea_since_2008['temp_differences'],'#5d35f3','Time since 2008','Differences in temperture','#0191f3','#2fec0e')

plt.show()
```

```python
def plot_nutrition(axes, x ,y, color, xlabel, ylabel,color1):
    axes.plot(x,y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color1)
    axes.tick_params('y',colors=color)
```

- Dark backgrounds are generally discouraged as they are less visible.
- If colors are important, consider using a colorblind-friendly style, such as "seaborn-colorblind" or "tableau-colorblind10". These are designed to retain color differences even when viewed by colorblind individuals.


```python
# Choosing a style
plt.style.use('ggplot')
fig, ax =plt.subplots()

ax.plot(starbucks_classic_pivot.index, starbucks_classic_pivot[' Sugars (g)'],color= '#07b0f2')
ax.plot(starbucks_classic_pivot.index, starbucks_classic_pivot[' Protein (g) '],color= '#733cc2')

ax2=ax.twinx()
ax2.plot(starbucks_classic_pivot.index, starbucks_classic_pivot['Calories'], color='#4aa61f')

ax2.set_ylabel('Calories', color='#4aa61f')
ax.set_ylabel('Sugar (g)', color='#07b0f2')
ax.set_xlabel('Beverage types')
ax.tick_params('x', rotation=270)


plt.show()
```
