```python
def revenue_runtime_2016(axes, x, y, color, xlabel, ylabel):
    axes.plot(x,y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
```

```python
fig, ax = plt.subplots()

revenue_runtime_2016(ax,df2_2016.index, df2_2016['revenue'], 'green','date','Revenue')

ax2= ax.twinx()
revenue_runtime_2016(ax2,df2_2016.index, df2_2016['runtime'], 'pink','date','Runtime' )

ax.set_xticklabels(df2_2016.index, rotation=45)

plt.show()
```
